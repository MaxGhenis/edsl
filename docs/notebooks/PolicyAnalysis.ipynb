# Get the Transcript from UCSB

```python
from edsl import Scenario 
speach = Scenario.from_url(
    'https://www.presidency.ucsb.edu/documents/address-before-joint-session-the-congress-4')
```

# Extract the economic policies & return them in a list

```python
from edsl import QuestionFreeText, QuestionList
q = QuestionFreeText(question_name = "economic_policies", 
                     question_text = "Describe all the economic policies in this speach: {{ text }}."
                    )
q_list = QuestionList(question_name = "policy", 
                      question_text = """Here is summary of economic policies: 
                      '{{ economic_policies.answer }}'.
                      
                      Please split this text into a list of policies---not just topics. Perserve all text in the responses.  
                      """
                     )
```

# Run the questions against the text, using GPT4o as the model

```python
results = q.add_question(q_list).by(speach).run()
```

# Create a list of the policies for further analysis

```python
policy_list = results.select('policy').expand('policy').to_scenario_list()
policy_list
```

# Create Agent Personas

```python
from edsl import Agent 
a1 = Agent(name = "economist", 
          instruction = "You are an expert, non-partisan economist evaluating a policy proposal.")
a2 = Agent(name = "rube", 
          instruction = "You are superficial, low-information voter who does not think much about policy.")
```

## Analyze the list of policies with the different personas

```python
q_analysis = QuestionFreeText(question_name = "analyze", 
                     question_text = "Please analyze this proposed economic policy: {{ policy }}"
                    )
policy_results = q_analysis.by(policy_list).by([a1, a2]).run()
```

```python
policy_results.select('agent.agent_name', 'policy', 'answer.*').order_by('policy')
```

# Let's make a markdown report of each analyst's analysis

```python
(policy_results
 .select('agent.agent_name', 'policy', 'answer.*')
 .order_by('policy')
 .report()
)
```
